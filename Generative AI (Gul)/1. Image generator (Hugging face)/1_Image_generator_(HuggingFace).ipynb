```python
import torch
from diffusers import StableDiffusionPipeline

print("Loading model...")

# Setup Device
if torch.cuda.is_available():
    device = "cuda"
    dtype = torch.float16
else:
    device = "cpu"
    dtype = torch.float32

# Load Model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype
)
pipe = pipe.to(device)

# --- CONFIGURATION ---
# 1. The Magic Suffix: This is added to EVERYTHING you type.
style_suffix = ", highly detailed, 8k resolution, cinematic lighting, photorealistic, trending on artstation, sharp focus"

# 2. The Negative Prompt: Things we want to avoid.
neg_prompt = "ugly, blurry, low quality, pixelated, distorted, bad anatomy, text, watermark"

while True:
    print("\n" + "="*40)
    user_input = input("Enter your basic idea (e.g., 'a cyberpunk city'): ")

    if user_input.lower() == 'exit':
        break

    if not user_input.strip():
        continue

    # --- PROMPT MODIFICATION LOGIC ---
    final_prompt = user_input + style_suffix

    print(f"\nOriginal Input: {user_input}")
    print(f"Modified Prompt: {final_prompt}")
    print(f"Generating...")

    try:
        # We pass both the prompt and the negative_prompt
        image = pipe(
            prompt=final_prompt,
            negative_prompt=neg_prompt,
            height=512,
            width=512
        ).images[0]

        # Save file
        clean_name = user_input[:15].replace(' ', '_')
        file_name = f"{clean_name}.png"
        image.save(file_name)
        print(f"Saved as: {file_name}")

    except Exception as e:
        print(f"Error: {e}")
```
